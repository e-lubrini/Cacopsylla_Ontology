```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
## substitute string with key column to identify each document
key_id = ''
```

```python
## open data file
with open(os.path.join('occurrence_data.csv'), 'r') as f:
    df = pd.read_csv(f)
```

```python
## treat data in listed column(s) as numeric
numeric_columns = ['dateIdentified'] 
for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
df.head()
```

```python
from tools.visual import plot
# TODO associated database associatedreference
save_dir = os.path.join(os.getcwd(), 'imgs')
```

```python
x = 'dateIdentified'
y = 'sourceCategory'
hue = 'sourceCategory'
palette = sns.color_palette('tab10',n_colors=2)

plot(x=x, y=y,
    hue=hue, data=df,
    type='violin',
    #split={hue:2},
    drop_na=[x],
    xlabel='date', ylabel='sourceCategory',
    huelabel='source',
    sort_alpha=[hue],
    swarm=True,
    #vlines=[
    #    {'x':1950,'ymin':0,'ymax':1,'gid':'copyright', 'c':palette[1]},
    #    {'x':2020,'ymin':0,'ymax':1,'gid':'last publication', 'c':'black'}
    #    ],
    save_dir = save_dir,
    )
```

```python
x = 'dateIdentified'
y = 'sourceCategory'
hue = 'sourceCategory'
style= 'sourceCategory'

plot(x=x, y=y, hue=hue, style=style, data=df,
    type='scatter',
    #split={hue:4,style:3,y:7}, #slice_top={y:5},
    legend_position='out',
    xlabel='date', ylabel='database',
    sort_alpha=[hue],
    save_dir = save_dir,
    )
```

```python
x = 'dateIdentified'
hue = 'PDF'
y = 'language'
palette = sns.color_palette('tab10',n_colors=10)

plot(x=x, y=y, hue=hue, data=df,
    type='violin',
    legend_position='lower left',
    slice_top={y:3},
    drop_doubles=[key_id], drop_na=[x],
    xlabel='date', ylabel='language', huelabel='pdf availability',
    #vlines=[
    #    {'x':1950,'ymin':0,'ymax':1,'gid':'copyright', 'c':palette[3]},
    #    {'x':2020,'ymin':0,'ymax':1,'gid':'last publication', 'c':'black'}
    #    ],
    swarm=True,
    save_dir = save_dir,
    )
# German: copyright from 1950
# English: a lot of info in unvailable documents
```

```python
x = 'dateIdentified'
hue = 'availability'
y = 'language'

plot(x=x, y=y, hue=hue, data=df,
    type='violin',
    legend_position='lower left',
    split={hue:2,y:2}, 
    xlabel='date', ylabel='language', huelabel='online availability',
    save_dir = save_dir,
    )
```
